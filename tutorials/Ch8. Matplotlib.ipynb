## Introduction

Matplotlib is a library for producing publication-quality figures. mpl (for short) was designed from the beginning to serve two purposes: 

 1. allow for interactive, cross-platform control of figures and plots
 2. make it easy to produce static raster or vector graphics files without the need for any GUIs. 
 
Furthermore, mpl -- much like Python itself -- gives the developer complete control over the appearance of their plots, while still being very usable through a powerful defaults system.

## Getting Started

In this tutorial, we'll use the following import statements. These abbreviations are semi-standardized, and most tutorials, other scientific python code that you'll find elsewhere will use them as well.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Prepare Data

```python
X = np.linspace (0, 4*np.pi, 100)
Y1 = np.sin(X)
Y2 = np.cos(X)
```

## Draw Plot

```python
# draw sine plot
fig, ax = plt.subplots() #no argument in subplots mean just one plot in figure
ax.plot(X,Y1) #plot X, Y1 in 'ax' axis (subplot)
plt.show()
```

```python
# draw cosine plot
fig, ax = plt.subplots()
ax.plot(X,Y2)
plt.show()
```

As you can see the function ``subplots()`` above generates two outputs: ``figure`` (fig) and ``axes`` (ax). Figure is the overall window/page that can contain multiple ``Axes``. 

<img src="https://raw.githubusercontent.com/matplotlib/AnatomyOfMatplotlib/master/images/figure_axes_axis_labeled.png">


# Axes
``Axes`` is effectively the area that we plot data on and any ticks/labels/etc associated with it.  Usually we'll set up an Axes with a call to ``subplot`` (which places Axes on a regular grid), so in most cases, ``Axes`` and ``Subplot`` are synonymous.  

```python
# plot two graphs on one axis (subplots)
fig, ax = plt.subplots()
ax.plot(X,Y1)
ax.plot(X,Y2)
plt.show()
```

```python
# plot two graphs on two axes (subplots)
fig, (ax1, ax2) = plt.subplots(2,1) #2 by 1 format
ax1.plot(X,Y1, color='C1')
ax2.plot(X,Y2, color='C0')
plt.show()
```

```python
# plot two graphs on two axes (subplots) 
fig, (ax1, ax2) = plt.subplots(1,2) #1 by 2 format
ax1.plot(X,Y1, color='C1')
ax2.plot(X,Y2, color='C0')
plt.show()
```

or we can also create ``figure`` first then add ``axis``

```python
fig = plt.figure()
ax = fig.add_subplot(111) # We'll explain the "111" later. Basically, 1 row and 1 column.
ax.plot(X,Y1)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211) # We'll explain the "111" later. Basically, it means  "2x1 grid, first subplot"
ax1.plot(X,Y1)

ax2 = fig.add_subplot(212) # "2x1 grid, second subplot"
ax2.plot(X,Y2)

plt.show()
```

When we want to change something about each axis, we use ``set_<something>``. and control a particular option. To demonstrate this (and as an example of IPython's tab-completion), try typing `ax.set_` in a code cell, then hit the `<Tab>` key.  You'll see a long list of `Axes` methods that start with `set`.

```python
fig, ax = plt.subplots()
ax.plot(X,Y1)

ax.set_title('A Different Example Axes Title')

ax.set_xlim([0, 10])
ax.set_ylim([0, 1])

ax.set_ylabel('Y-Axis (changed)')
ax.set_xlabel('X-Axis (changed)')

ax.set_xticks(range(0, 11, 2))
ax.set_xticklabels(['zero', 'two', 'four', 'six', 'eight', 'ten'])

plt.show()
```

```python
## or you can do this way
fig, ax = plt.subplots()
ax.plot(X,Y1)

ax.set(xlim= [0, 10], ylim=[0, 1], title ='A Different Example Axes Title', 
       ylabel='Y-Axis (changed)',xlabel= 'X-Axis (changed)',
      xticks =range(0, 11, 2), xticklabels=['zero', 'two', 'four', 'six', 'eight', 'ten'] )
plt.show()
```

```python
# you can also use ticklocators API for xticks and yticks setting

from matplotlib import ticker
fig, ax = plt.subplots()
ax.plot(X,Y1)

ax.set_title('A Different Example Axes Title')

ax.set_xlim([0, 10])
ax.set_ylim([0, 1])

ax.set_ylabel('Y-Axis (changed)')
ax.set_xlabel('X-Axis (changed)')

ax.xaxis.set_major_locator(ticker.MultipleLocator(2.0))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.5))

plt.show()
```

## modify graphs 
You can modify colors, markers, styles of graphs by passing arguments in ``.plot``

```python
# this is default plot
fig, ax = plt.subplots() 
ax.plot(X,Y1)
plt.show()
```

```python
# change linecolor
fig, ax = plt.subplots() 
ax.plot(X,Y1, color='black')
plt.show()
```

```python
# change line style
fig, ax = plt.subplots() 
ax.plot(X,Y1, linestyle='--')
plt.show()
```

```python
# change linewidth
fig, ax = plt.subplots() 
ax.plot(X,Y1, linewidth=5)
plt.show()
```

```python
# change marker
fig, ax = plt.subplots() 
ax.plot(X,Y1, marker='o')
plt.show()
```

## Matplotlib offers several kinds of plots!
Also see [galleries](https://matplotlib.org/gallery/index.html)

```python
# scatter plot
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)
fig,ax = plt.subplots()

ax.scatter(X,Y)
plt.show()
```

```python
# bar plot
X = np.arange(10)
Y = np.random.uniform(0,10,10)
fig,ax = plt.subplots()
ax.bar(X,Y)
plt.show()
```

```python
#histogram
Z = np.random.normal(0,1,100)
fig,ax = plt.subplots()
ax.hist(Z)
plt.show()
```

## Miscellenous

```python
## text and ornaments
X = np.linspace(0, 10, 100)
Y = np.cos(X)
fig, ax = plt.subplots()

ax.plot(X,Y)

ax.fill_between(x=[0,5],y1=-1, y2=1, color='black', alpha = 0.2)
ax.text(0, -1, "Period $\phi$")

plt.show()
```
